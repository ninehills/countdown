```python
import matplotlib.pyplot as plt
import numpy as np

# 定义模型名称和对应的COT准确率
models = ['deepseek-r1', 'Qwen2.5-1.5B-Instruct',
          'DeepSeek-R1-Distill-Qwen-1.5B', 'Qwen2.5-1.5B-simple-sft', 'Qwen2.5-1.5B-distill-sft']

accuracies = [82.0, 2.0, 0.0, 4.0, 44.0]

# 创建颜色映射
colors = ['#1f77b4'] * 3 + ['#ff7f0e'] * 2  # 最后两个模型使用不同颜色

# 创建图表
plt.figure(figsize=(12, 6))
bars = plt.bar(models, accuracies, color=colors)

# 设置标题和标签
plt.title('Accuracy', fontsize=14)
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)

# 旋转x轴标签以防重叠
plt.xticks(rotation=45, ha='right')

# 在柱子上方显示具体数值
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}%', ha='center', va='bottom')

# 调整布局以确保标签完全显示
plt.tight_layout()

# 显示图表
plt.show()
```
